# Task 03: Geospatial Analysis

## ➣ Objective

- To visualize restaurant locations using latitude and longitude data and analyze their distribution across cities and countries.

## ➣ Tools Used

- Python  
- Pandas  
- Matplotlib  
- Seaborn  
- Folium  
- Google Colab  

## Step 1: Import Libraries & Load Dataset

- In this step, we import necessary libraries and load the dataset.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import folium

df = pd.read_csv('Dataset.csv')
df.head()
```

## Step 2: Understanding Location Data

- Here, we explore latitude and longitude columns used for geospatial analysis.

```python
df[['Latitude', 'Longitude']].head()
```

## Step 3: Visualizing Restaurant Locations on Map

- This map visualizes restaurant locations using latitude and longitude information.

```python
# Create base map
restaurant_map = folium.Map(
    location=[df['Latitude'].mean(), df['Longitude'].mean()],
    zoom_start=2
)

# Add first 100 restaurant locations
for index, row in df.head(100).iterrows():
    folium.Marker(
        [row['Latitude'], row['Longitude']],
        popup=row['Restaurant Name']
    ).add_to(restaurant_map)

restaurant_map
```

## Step 4: Distribution of Restaurants Across Cities

- This chart shows cities with the highest number of restaurants.

```python
plt.figure(figsize=(12,5))

sns.countplot(
    x='City',
    data=df,
    order=df['City'].value_counts().head(10).index
)

plt.title("Top Cities with Most Restaurants")
plt.xticks(rotation=45)
plt.show()
```

## Step 5: Distribution of Restaurants Across Countries

- This chart shows the distribution of restaurants across different countries.

```python
plt.figure(figsize=(10,5))

sns.countplot(
    x='Country Code',
    data=df,
    order=df['Country Code'].value_counts().index
)

plt.title("Restaurant Distribution by Country")
plt.show()
```

## Step 6: Correlation Between Location and Rating

- This scatter plot analyzes the relationship between restaurant location and aggregate rating.

```python
plt.figure(figsize=(10,6))

sns.scatterplot(
    x='Longitude',
    y='Latitude',
    hue='Aggregate rating',
    data=df
)

plt.title("Restaurant Location vs Aggregate Rating")
plt.show()
```

## ➣ Conclusion

- Visualized restaurant locations using latitude and longitude  
- Identified cities with the highest number of restaurants  
- Analyzed restaurant distribution across countries  
- Explored the relationship between location and restaurant ratings  

This analysis helps understand geographical patterns in restaurant data.
